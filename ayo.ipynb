```python
import pandas as pd
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo

heart_disease = fetch_ucirepo(id=45)

X = heart_disease.data.features
y = heart_disease.data.targets

df = pd.concat([X, y], axis=1)

print(df.isnull().sum())

df = df.dropna()

print("DATA BERSIHH >_<:", df.shape)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['age'], df['trestbps'], linestyle='dashed', marker='o')
plt.title('Hubungan Umur vs Tekanan Darah')
plt.xlabel('Umur')
plt.ylabel('Tekanan Darah')
plt.grid(True)
plt.show()
```

```python
df_sorted = df.sort_values('age')

plt.figure(figsize=(10,5))
plt.plot(df_sorted['age'], df_sorted['trestbps'], marker='o', linestyle='-', markersize=4)

plt.title('Tren Tekanan Darah Berdasarkan Umur')
plt.xlabel('Umur')
plt.ylabel('Tekanan Darah')
plt.grid(True)
plt.show()
```

```python
# jumlah pasien berdasarkan target (0 = sehat, 1 = sakit)
data_bar = df.groupby('num').size()

plt.figure(figsize=(6,4))
plt.bar(data_bar.index, data_bar.values)
plt.title('Jumlah Pasien Berdasarkan Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Jumlah')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['chol'])
plt.title('Scatter Umur vs Kolesterol')
plt.xlabel('Umur')
plt.ylabel('Kolesterol')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(data_bar.values, labels=data_bar.index)
plt.title('Persentase Diagnosis Penyakit Jantung')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=10)
plt.title('Distribusi Umur Pasien')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')
plt.show()
```
